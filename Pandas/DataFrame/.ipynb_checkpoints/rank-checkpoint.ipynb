```python
import pandas as pd
```

```python
nba=pd.read_csv('nba.csv').dropna(how='all')
nba.head()
```

```python
nba["Salary"]=nba["Salary"].fillna(0).astype(int)
nba.head()
```

```python
nba["SalaryRank"]=nba["Salary"].rank(ascending=False).astype(int)
nba.head()
```

```python
nba.sort_values("Salary",ascending=False,inplace=True)
nba.head(20)
```
