```python
import pandas as pd
```

```python
bond=pd.read_csv('jamesbond.csv',index_col='Film')
bond.sort_index()
bond.head()
```

```python
bond["Actor"].head()
```

```python
mask=bond["Actor"]=="Sean Connery"
bond[mask]["Actor"]="Sir Connery"
```
