```python
import numpy as np
import matplotlib.pyplot as pt
```

```python
x1=np.linspace(0.0,5.0,100) #linspace(start,end,equally divied into segments)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1) #cos(2*pi*x1)*e^(-x1)
y2=np.cos(2*np.pi*x2)

#subplot 1

pt.subplot(2,1,1) #subplot(nRows,nColumn,index)
pt.plot(x1,y1)
pt.title('sub-plot 1')
pt.xlabel('X1')
pt.ylabel('Y1')
pt.show()

pt.subplot(2,1,2) #subplot(nRows,nColumn,index) here we change the index for second plot
pt.plot(x2,y2)
pt.title('sub-plot 2')
pt.xlabel('X2')
pt.ylabel('Y2')
pt.show()
```
