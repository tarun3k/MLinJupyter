```python
import matplotlib.pyplot as pt
import csv
```

```python
x=[]
y=[]

with open('plot.txt','r') as csvfile:
    plots=csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])

pt.plot(x,y,label='From-Text')
x=[]
y=[]
with open('plot.csv','r') as csvfile:
    plots=csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])

pt.plot(x,y,label='From-CSV')
pt.xlabel('x')
pt.ylabel('y')
pt.title('From file')

pt.legend()
pt.show()
```
