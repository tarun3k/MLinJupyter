```python
import matplotlib.pyplot as pt
import matplotlib.animation as animation
from matplotlib import style
```

```python
pt.style.use('fivethirtyeight')
fig=pt.figure()
axis1=fig.add_subplot(111)
def animate(p):
    x=[]
    y=[]
    full_data=open('plot.txt','r').read()
    line_data=full_data.split('\n')
    for line in line_data:
        if(len(line)>2):
            x1,y1=line.split(',')
            x.append(x1)
            y.append(y1)
        axis1.clear()
        axis1.plot(x,y)

anime_data=animation.FuncAnimation(fig,animate,interval=500)
pt.show()
            
```
