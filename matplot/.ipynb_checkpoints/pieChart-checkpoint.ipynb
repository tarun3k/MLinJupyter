```python
import numpy as np
import matplotlib.pyplot as pt
```

```python
act=['eat','sleep','gym','study','coding']
slices=[1,2,1,4,3]
color=['r','g','m','b','orange']

pt.pie(slices,labels=act,colors=color,startangle=90,shadow=True,explode=[.2,0.9,0,0,0],autopct='%1.1f%%')
pt.legend();
pt.show()
```
