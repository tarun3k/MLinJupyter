```python
import numpy as np
import matplotlib.pyplot as pt
```

```python
#x is same for both y
x=np.arange(0.0,2*np.pi,0.1)
y1=np.sin(x)
y2=np.cos(x)
#plot two plot on one plane
pt.plot(x,y1,label='sin')
pt.plot(x,y2,label='cos')

pt.xlabel('x-axis')
pt.ylabel('y-axis')
#to show labels
pt.legend()
pt.show()
```
